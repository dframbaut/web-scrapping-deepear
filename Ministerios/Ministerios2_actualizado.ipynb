**Ministerio de hacienda**

Descargar el documento Ministerio de Hacienda (1).xlsx

```python
from google.colab import files
import pandas as pd

archivo_subido_hacienda= files.upload()

# Leer el archivo Excel
nombre_archivo_h = list(archivo_subido_hacienda.keys())[0]  # Obtiene el nombre del archivo subido
df = pd.read_excel(nombre_archivo_h)

# Mostrar las primeras filas del dataframe
print(df.head())
```

```python
import pandas as pd

# Ruta del archivo de Excel
archivo_excel_h = "Ministerio de hacienda (1).xlsx"

# Leer la hoja de cálculo
hojas_h = pd.ExcelFile(archivo_excel_h)

# Cargar la hoja inicial
df_h = hojas_h.parse(sheet_name=hojas_h.sheet_names[0])

# Convertir nombres de columnas a formato estándar (sin espacios iniciales o finales)
df_h.columns = df_h.columns.str.strip()

# Definir las columnas que contienen 'sí' o 'no'
columnas_si_no = ['es_reglamentario', 'tiene_costo', 'adopta_algun_trámite', 'deroga']

# Verificar y procesar las columnas específicas
for columna in columnas_si_no:
    if columna in df_h.columns:
        # Convertir a minúsculas
        df_h[columna] = df_h[columna].str.lower()

        # Reemplazar 'sí' por True y 'no' por False
        df_h[columna] = df_h[columna].replace({'si': True, 'no': False}, regex=True)

# Ajustar la columna 'entidades' al formato JSON
if 'entidades' in df_h.columns:
    df_h['entidades'] = df_h['entidades'].apply(lambda x: f'{{"entidades": ["{x}"]}}')

# Completar valores nulos en la columna 'norma_modifica' con 'No definido'
if 'norma_modifica' in df_h.columns:
    df_h['norma_modifica'] = df_h['norma_modifica'].fillna('No definido')

# Mostrar las primeras filas del DataFrame modificado
print(df_h.head())

# Guardar el DataFrame procesado en un nuevo archivo Excel
df_h.to_excel("modificado_Agenda_Regulatoria_2025.xlsx", index=False)

print("El archivo modificado se ha guardado como 'modificado_Agenda_Regulatoria_2025.xlsx'")
```
